# loc and iloc

```python
import seaborn as sns
import pandas as pd
iris_data = sns.load_dataset('iris')
df = iris_data.head(5).copy()
```

```python
df
```

##### loc works with label values and iloc works with integer values. the format is [rows,columns]

```python
df = pd.concat([df,iris_data.iloc[50:55]])
df = pd.concat([df,iris_data.iloc[100:105]])
```

```python
 df = df.reset_index(drop = True)
```

## Select first row and all columns

```python
df.loc[0, :]
```

```python
df.loc[[0,1,2],:]
```

```python
df.loc[[0,10],:]
```

```python
df.loc[0:8,:]
```

#### note that loc includes the ranges

```python
df.loc[:,'sepal_length']
```

```python
df.loc[:,['sepal_length','species']]
```

```python
df.loc[:,'sepal_length':'petal_width']
```

```python
df.loc[:,'petal_length':'petal_width']
```

```python
df.loc[:,['petal_length','petal_width']]
```

```python
df.loc[df.species == 'setosa']
```

```python
df.loc[df.species == 'setosa','petal_length':'petal_width']
```

###### note that if ou are using : for a column range labels you dont have to use double brackets. Only when the columns are not adjacet and you specify them using commas you should double bracket it

### lets move on to iloc

```python
df.iloc[:,0:2]
```

```python
df.iloc[1:4,0:2]
```

```python
df.iloc[0:2,:]
```

### is same as

```python
df[0:2]
```

#### but as a standard practice use loc or iloc

pandas by default uses iloc if not specified
